```python
from pyspark.sql import SparkSession
import pandas as pd
from pyspark.sql.functions import lit

spark = SparkSession.builder.appName("SchemaHandling").getOrCreate()

data = {
    "ProductID": [201, 202, 203, 204],
    "ProductName": ["Laptop", "Smartphone", "Tablet", "Smartwatch"],
    "Category": ["Electronics", "Electronics", "Electronics", "Wearables"],
    "Price": [1200, 800, 500, 300],
    "Stock": [50, 200, 150, 100]
}

df = pd.DataFrame(data)
csv_file = "/tmp/products.csv"
df.to_csv(csv_file, index=False)

product_df = spark.read.option("header", "true").option("inferSchema", "true").csv(csv_file)

product_df.printSchema()
product_df.show()

updated_product_df = product_df.withColumn("Discount", lit(None).cast("double"))
updated_product_df.printSchema()
updated_product_df.show()

product_df.createOrReplaceTempView("products")

spark.sql("SELECT * FROM products").show()
spark.sql("SELECT ProductName, Price FROM products WHERE Stock < 100").show()

```

```python
from pyspark.sql import SparkSession
import pandas as pd
from pyspark.sql.functions import lit

spark = SparkSession.builder.appName("SchemaHandling").getOrCreate()

data = {
    "VehicleID": [1, 2, 3, 4],
    "Brand": ["Toyota", "Honda", "Ford", "Tesla"],
    "Model": ["Corolla", "Civic", "Focus", "Model 3"],
    "Year": [2018, 2019, 2020, 2021],
    "Price": [20000, 22000, 25000, 35000]
}

df = pd.DataFrame(data)
csv_file = "/tmp/vehicles.csv"
df.to_csv(csv_file, index=False)

vehicle_df = spark.read.option("header", "true").option("inferSchema", "true").csv(csv_file)

vehicle_df.printSchema()
vehicle_df.show()

updated_vehicle_df = vehicle_df.withColumn("Mileage", lit(None).cast("integer"))
updated_vehicle_df.printSchema()
updated_vehicle_df.show()

vehicle_df.createOrReplaceTempView("vehicles")

spark.sql("SELECT * FROM vehicles").show()
spark.sql("SELECT Brand, Model, Price FROM vehicles WHERE Price > 25000").show()




```

```python
from pyspark.sql import SparkSession
import pandas as pd
from pyspark.sql.functions import lit

spark = SparkSession.builder.appName("SchemaHandling").getOrCreate()

data = {
    "CourseID": [101, 102, 103, 104],
    "CourseName": ["Python Basics", "Data Science", "Web Development", "Machine Learning"],
    "Instructor": ["Dhoni", "Virat", "Raina", "Rohit"],
    "Duration": [4, 8, 6, 10],
    "StudentsEnrolled": [150, 200, 180, 220]
}

df = pd.DataFrame(data)
csv_file = "/tmp/courses.csv"
df.to_csv(csv_file, index=False)

course_df = spark.read.option("header", "true").option("inferSchema", "true").csv(csv_file)

course_df.printSchema()
course_df.show()

updated_course_df = course_df.withColumn("Rating", lit(None).cast("double"))
updated_course_df.printSchema()
updated_course_df.show()

course_df.createOrReplaceTempView("courses")

spark.sql("SELECT * FROM courses").show()
spark.sql("SELECT CourseName, Instructor FROM courses WHERE StudentsEnrolled > 180").show()
```

```python
from pyspark.sql import SparkSession
import pandas as pd
from pyspark.sql.functions import lit

spark = SparkSession.builder.appName("SchemaHandling").getOrCreate()

data = {
    "TransactionID": [1001, 1002, 1003, 1004],
    "CustomerName": ["Alice", "Bob", "Charlie", "David"],
    "Product": ["Milk", "Bread", "Eggs", "Butter"],
    "Quantity": [2, 1, 1, 3],
    "TotalPrice": [10, 5, 6, 15]
}

df = pd.DataFrame(data)
csv_file = "/tmp/supermarket_sales.csv"
df.to_csv(csv_file, index=False)

sales_df = spark.read.option("header", "true").option("inferSchema", "true").csv(csv_file)

sales_df.printSchema()
sales_df.show()

updated_sales_df = sales_df.withColumn("DiscountApplied", lit(False).cast("boolean"))
updated_sales_df.printSchema()
updated_sales_df.show()

sales_df.createOrReplaceTempView("sales")

spark.sql("SELECT * FROM sales").show()
spark.sql("SELECT CustomerName, Product, TotalPrice FROM sales WHERE TotalPrice > 10").show()
```
